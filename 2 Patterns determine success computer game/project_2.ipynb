**В столбце User_Score** кроме NaN eсть **<span style="color:red">tbd</span>** - аббревиатура To Be Determined (не определен).   
tbd отличается от NaN. Если пропуск появился из-за "ошибки" заполнения данных (то есть рейтинг есть, но его забыли записать), а tbd означает что данные еще обрабатываются - идет процесс определения рейтинга). Исходя из этого:    
**<span style="color:blue">Необходимо</span>** tbd - **заменить на -1** и в расчет не брать(задавать условие >= 0), но сначала перевести данные в float

**<span style="color:blue">После этого</span> **перевести данные** к 100 бальной шкале.   
**<span style="color:blue">Далее</span> изменить тип данных** на int.

## Проверка гипотез

### Гипотеза № 1

Формулируем нулевую гипотезу:   
H₀ - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые   
Альтернативная гипотеза:   
H₁ - Средние пользовательские рейтинги платформ Xbox One и PC различаются   
Для проверки гипотезы используем метод: scipy.stats.ttest_ind (array1, array2, equal_var), а так как дисперсии выборок не равны, то equal_var=False   
критический уровень статистической значимости определяем 0.01 (1%)   

```python
alpha = 0.01
data_XOne = data.query('platform == "XOne" and user_score >= 0')
data_PC = data.query('platform == "PC" and user_score >= 0')
results = st.ttest_ind(data_XOne['user_score'], data_PC['user_score'], equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:**
Гипотеза, что cредние пользовательские рейтинги платформ Xbox One и PC одинаковые - **отвергается**

### Гипотеза № 2

Формулируем нулевую гипотезу:   
H₀ - Средние пользовательские рейтинги жанров Action и Sports одинаковые   
Альтернативная гипотеза:   
H₁ - Средние пользовательские рейтинги жанров Action и Sports различаются   
Для проверки гипотезы используем метод: scipy.stats.ttest_ind (array1, array2, equal_var), а так как дисперсии выборок не равны, то equal_var=False   
критический уровень статистической значимости определяем 0.01 (1%)   

```python
data['genre'].value_counts()
```

```python
alpha = 0.01
data_Action = data.query('genre == "Action" and user_score >= 0')
data_Sports = data.query('genre == "Sports" and user_score >= 0')
results = st.ttest_ind(data_Action['user_score'], data_Sports['user_score'], equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
